# **Abstraction:**

Cho biết a,k,b,m,n, tìm số nguyên x thỏa: x(a+b)-a(x//k)-b(x//m) >= n

# **Pattern Recognition:**

Thuật toán: Vét cạn ( tối ưu )

Đặc điểm nhận dạng: Bài toán yêu cầu phải duyệt từng ngày  vét cạn khá hợp lí

# **Algorithm Design :**

Sử dụng thuật toán vét cạn. Tuy nhiên nếu duyệt qua hết tất cả các ngày thì sẽ rất tốn thời gian, không đáp ứng được thời gian của đề bài với đầu vào dữ liệu lớn nên chúng ta sẽ tối ưu lại thuật toán vét cạn trong bài toán này.

Thay vì xuất phát từ 0 thì ta sẽ xuất phát từ 1 ngày nào đó mà gần với kết quả nhất  giảm số lần duyệt của thuật toán mà vẫn đảm bảo kết quả. Cụ thể:

- Ta có n là tổng số cây cần đốn hạ, a và b lần lượt là số lượng cây mà mỗi đội đốn trong 1 ngày  Nếu 2 đội làm việc mỗi ngày, không có ngày nghỉ, thì cần tối thiểu y = n//(a+b) ngày để đốn hết số cây --> Ta có thể bắt đầu duyệt từ đây.

- Tuy nhiên khi thực hành thì vẫn chưa đáp ứng đc thời gian của bài  Duyệt từ y= n//(a+b) là đúng nhưng vẫn chưa phải là con số tối ưu nhất --> Tính toán lại x:
   - Xét số cây còn lại ở ngày thứ x. Nếu số cây còn lại lớn hơn tổng cây của 2 đội đốn mỗi ngày  cần thêm tối thiểu z= ( cây còn lại // tổng cây 2 đội mỗi ngày) ngày nữa thì mới có thể hoàn thành --> Vậy z = x + y là con số tối ưu nhất.

# **# Programming:**



```
b=list(map(int,input().split())) 
day=b[4]//(b[0]+b[2]) # ngày gần nhất
x=day//b[1] 
y=day//b[3]
tong=day*(b[0]+b[2])-x*b[0]-y*b[2] # số cây đang có của ngày đó 
con_lai = b[4]- tong # số cây còn lại
while con_lai >= (b[0]+b[2]): # nếu cây còn lại lớn hơn or bằng tổng số cây 2 đội thì cần tối thiểu (cây còn lại // tổng cây 2 đôi) ngày nữa 
  con_lai = con_lai // (b[0]+b[2]) # tính lại ngày và tổng
  day+=con_lai
  x=day//b[1] 
  y=day//b[3]
  tong=day*(b[0]+b[2])-x*b[0]-y*b[2]
  con_lai = b[4]- tong
while tong < b[4]: 
  day=day+1
  if day % b[3]==0 and day % b[1]==0:
    tong+=0
  elif (day % b[3]==0):
    tong+=b[0]
  elif (day % b[1]==0):
    tong+=b[2]
  else:
    tong=tong+b[2]+b[0]
print(day)
```
